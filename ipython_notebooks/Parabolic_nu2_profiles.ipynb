```python
plt.plot(r, splines['b_z'](r))
plt.title(r'$B_{z}$', fontsize=20)
plt.show()
```

```python
plt.plot(r, splines['pprime'](r))
plt.title(r"$p'$", fontsize=20)
plt.show()
```
