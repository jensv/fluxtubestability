```python
#levels = [-240, -1, -0.1, 0.1,  1, 80]
levels = None
test_data = np.load('../../output/2015-03-27-21-53/meshes.npz')
test_lambda = test_data['lambda_a_mesh']
test_k = test_data['k_a_mesh']
test_delta = test_data['delta_m_0'] 

#print(test_delta[test_delta > 1])
test_delta = np.clip(test_delta, -1.5, 1)

#print(np.sum(test_delta < -1))
#test_delta = test_delta/np.nanmax(np.abs(test_delta))

colors = sns.light_palette('orange', n_colors=6, reverse=True, as_cmap=True)

if levels:
    plt.contourf(test_lambda, test_k, test_delta, cmap=colors, levels=levels)
else:
    plt.contourf(test_lambda, test_k, test_delta, cmap=colors)
    
cbar = plt.colorbar(label=r'$\delta$')
cbar.set_label(label=r'$\delta$', size=45, rotation=0, labelpad=30)

if levels:
    plt.contour(test_lambda, test_k, test_delta, colors='grey', levels=levels)
else:
    plt.contour(test_lambda, test_k, test_delta, colors='grey')
    
plt.ylim(0.01, 1.5)
plt.xlim(0.01, 3.0)
axes = plt.gca()
axes.set_axis_bgcolor(sns.xkcd_rgb['grey'])
plt.setp(axes.get_xticklabels(), fontsize=40)
plt.setp(axes.get_yticklabels(), fontsize=40)
plt.ylabel(r'$\bar{k}$', fontsize=45, rotation='horizontal', labelpad=30)
plt.xlabel(r'$\bar{\lambda}$', fontsize=45)
cbar.ax.tick_params(labelsize=40)
sns.despine(ax=axes)
plt.tight_layout()
```
