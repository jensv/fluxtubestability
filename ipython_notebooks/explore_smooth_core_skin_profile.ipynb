## Look at how $\epsilon$ varies $j_{z}$, $B_{\theta}$, $p$ profiles

### Create Profiles

```python
ep_08_profile = es.UnitlessSmoothedCoreSkin(epsilon=0.8, lambda_bar=1.0, k_bar=1.)
ep_09_profile = es.UnitlessSmoothedCoreSkin(epsilon=0.9, lambda_bar=1.0, k_bar=1.)
ep_10_profile = es.UnitlessSmoothedCoreSkin(epsilon=1.0, lambda_bar=1.0, k_bar=1.)
ep_15_profile = es.UnitlessSmoothedCoreSkin(epsilon=1.5, lambda_bar=1.0, k_bar=1.)
```

### Get Splines

```python
ep_08_splines = ep_08_profile.get_splines()
ep_09_splines = ep_09_profile.get_splines()
ep_10_splines = ep_10_profile.get_splines()
ep_15_splines = ep_15_profile.get_splines()
```

```python
epsilon_labels = [0.8, 0.9, 1.0, 1.5]
r = np.linspace(0, 1, 1000)
```

### Plot Axial Current

```python
fig, axs = plt.subplots(1,4, figsize=(20,5), sharey=True)
axs[0].plot(r, ep_08_splines['j_z'](r))
axs[1].plot(r, ep_09_splines['j_z'](r))
axs[2].plot(r, ep_10_splines['j_z'](r))
axs[3].plot(r, ep_15_splines['j_z'](r))
    
axs[0].set_ylabel(r'$\tilde{j}_{z}$', fontsize=25)
for i, axis in enumerate(axs):
    axis.set_xlabel(r'$r$', fontsize=25)
    epsilon = epsilon_labels[i]
    axis.set_title(r'$\epsilon = $' + str(epsilon), fontsize=25)
```

### Plot Azimuthal Field

```python
fig, axs = plt.subplots(1,4, figsize=(20,5), sharey=True)
axs[0].plot(r, ep_08_splines['b_theta'](r))
axs[1].plot(r, ep_09_splines['b_theta'](r))
axs[2].plot(r, ep_10_splines['b_theta'](r))
axs[3].plot(r, ep_15_splines['b_theta'](r))
    
axs[0].set_ylabel(r'$\tilde{B}_{\theta}$', fontsize=25)
for i, axis in enumerate(axs):
    axis.set_xlabel(r'$r$', fontsize=25)
    epsilon = epsilon_labels[i]
    axis.set_title(r'$\epsilon = $' + str(epsilon), fontsize=25)
```

## $\beta$ seems to be high. How does $\beta$ change with larger skin regions? 

```python
beta_default_profile = es.UnitlessSmoothedCoreSkin(beta_0=100., epsilon=0.7, lambda_bar=1.0, k_bar=1., j_0=1.)
beta_elongated_profile = es.UnitlessSmoothedCoreSkin(beta_0=0.5, epsilon=0.7, lambda_bar=1.0, k_bar=1., j_0=1., core_radius_norm=0.1, transition_width_norm=0.1, skin_width_norm=0.4)
beta_elongated_inverted_profile = es.UnitlessSmoothedCoreSkin(beta_0=0.5, epsilon=1.2, lambda_bar=1.0, k_bar=1., j_0=1., core_radius_norm=0.1, transition_width_norm=0.1, skin_width_norm=0.4)
beta_low_lambda_profile = es.UnitlessSmoothedCoreSkin(beta_0=0.5, epsilon=0.7, lambda_bar=0.1, k_bar=1., j_0=1.)
```

```python
beta_default_beta_spl = beta_default_profile.get_beta(beta_default_profile.r)
beta_elongated_beta_spl = beta_elongated_profile.get_beta(beta_elongated_profile.r)
beta_elongated_inverted_beta_spl = beta_elongated_inverted_profile.get_beta(beta_elongated_inverted_profile.r)
beta_low_lambda_beta_spl = beta_low_lambda_profile.get_beta(beta_low_lambda_profile.r)
```

```python
beta_labels = ['default', 'elongated', 'elongated inverted', 'low lambda']
```

```python
fig, axs = plt.subplots(1, 4, figsize=(20,5), sharey=True)
axs[0].plot(r, beta_default_beta_spl(r))
axs[1].plot(r, beta_elongated_beta_spl(r))
axs[2].plot(r, beta_elongated_inverted_beta_spl(r))
axs[3].plot(r, beta_low_lambda_beta_spl(r))

    
axs[0].set_ylabel(r'$\beta$', fontsize=25)
for i, axis in enumerate(axs):
    axis.set_xlabel(r'$r$', fontsize=25)
    beta_label = beta_labels[i]
    axis.set_title(r'' + str(beta_label), fontsize=25)
```
