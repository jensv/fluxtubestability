```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
sns.set_context('poster')
import numpy as np
```

I want to solve the ODE:
$$(f \xi)' - g \xi =0 $$

I can write the 2nd order ODE as two 1st order ODEs:
$$\xi' = \frac{f \xi'}{f} $$
$$(f \xi)' = g \xi $$

where
$$ f = \frac{r (krB_z + m B_{\theta})^2}{k^2r^2+m^2}$$


$$ g=\frac{2k^2r^2\beta_{0}}{k^2r^2+m^2} \frac{dP}{dr} + \frac{1}{r}(krB_{z} + m B_{\theta})^{2} \frac{k^{2}r^{2}+m^{2}-1}{k^{2}r^{2}+m^{2}} + \frac{2k^{2}r}{(k^{2}r^{2}+m^{2})^{2}}(k^{2}r^{2}B_{z}^{2}-m^{2}B_{\theta}^{2}) $$

when $m=0$

$$f = r B_z^2$$


$$g = \beta_0 2 \frac{dp}{dr}+ k^2 r B_z^2 + \frac{B_z^2}{r}$$

```python
def f(r):
    return r
def g(r):
    return -1 + r + 1./r
```

```python
def der(r, y):
    y_der = np.zeros(2)
    y_der[0] = y[1]/f(r)
    y_der[1] = g(r)*y[0]
    return y_der
```

```python
from scipy.integrate import ode
```

```python
integrator = ode(der)
integrator.set_integrator('lsoda')
r_init = 1E-3
init = [.3, .9]
integrator.set_initial_value(init, t=r_init)
r = np.linspace(r_init, 1., 1000000)
results = np.zeros((2, r.size))
results[:,0] = init
for i, position in enumerate(r[1:]):
    integrator.integrate(position)
    results[:, i+1] = integrator.y
```

```python
plt.plot(r, results[0])
```

```python
plt.plot(r, results[1])
```

The residual should be:
$$ f \xi'' + f' \xi' - g \xi = 0 $$

```python
dr = r[1] - r[0]
```

```python
f(r)*np.gradient(results[1]/f(r))/dr + np.gradient(f(r))/dr*results[1]/f(r) - g(r)*results[0]
```

```python
def f2(r):
    return 4.
def g2(r):
    return 5.
def der2(r, y):
    y_der = np.zeros(2)
    y_der[0] = y[1]/f2(r)
    y_der[1] = g2(r)*y[0]
    return y_der
```

```python
integrator = ode(der2)
integrator.set_integrator('lsoda')
init = [0., 1.]
integrator.set_initial_value(init, t=0.001)
r = np.linspace(0.001, 1., 100)
results = np.zeros((2, r.size))
results[:,0] = init
for i, position in enumerate(r[1:]):
    integrator.integrate(position)
    results[:, i+1] = integrator.y
```
