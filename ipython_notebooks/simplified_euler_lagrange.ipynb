I want to solve the ODE:
$$(f \xi)' - g \xi =0 $$

I can write the 2nd order ODE as two 1st order ODEs:
$$\xi' = \frac{f \xi'}{f} $$
$$(f \xi)' = g \xi $$

where
$$ f = \frac{r (krB_z + m B_{\theta})^2}{k^2r^2+m^2}$$


$$ g=\frac{k^2r^2\beta_{0}}{k^2r^2+m^2} \frac{dP}{dr} + \frac{1}{r}(krB_{z} + m B_{\theta})^{2} \frac{k^{2}r^{2}+m^{2}-1}{k^{2}r^{2}+m^{2}} + \frac{2k^{2}r}{(k^{2}r^{2}+m^{2})^{2}}(k^{2}r^{2}B_{z}^{2}-m^{2}B_{\theta}^{2}) $$

when $m=0$

$$f = r B_z^2$$

$$g = \beta_0 \frac{dp}{dr}+ k^2 r B_z^2 + \frac{B_z^2}{r}$$

In my profiles $B_z=1$ and since $j_z$ is constant in the core: $\frac{dp}{dr}=-c r$:

$$f = r$$
$$g = -\beta_0 c r+ k^2 r + \frac{1}{r}$$

Fo simplicity let's set everything equal to 1.

## Set up a simple f and g

```python
def f(r):
    return r
def g(r):
    return -r + 1./r
```

```python
def der(r, y):
    y_der = np.zeros(2)
    y_der[0] = y[1]/f(r)
    y_der[1] = g(r)*y[0]
    return y_der
```

```python
integrator = ode(der)
integrator.set_integrator('lsoda', rtol=1E-10)
r_init = 1E-3
init = [r_init, 1.*f(r_init)]
integrator.set_initial_value(init, t=r_init)
r = np.linspace(r_init, 1., 100)
results = np.zeros((2, r.size))
results[:,0] = init
for i, position in enumerate(r[1:]):
    integrator.integrate(position)
    results[:, i+1] = integrator.y
```

The residual should be:
$$ f \xi'' + f' \xi' - g \xi = 0 $$

I have to approximate the second derivative of $\xi$ with a second oder finite difference. 

```python
dr = r[1] - r[0]
residual = f(r)*np.gradient(results[1]/f(r))/dr + np.gradient(f(r))/dr*results[1]/f(r) - g(r)*results[0]
```

## The residual is small, larger on the edges due to first order derivative in $\xi'$ cacluation

```python
plt.semilogy(r, np.abs(residual))
plt.xlabel(r'r', fontsize=25)
plt.ylabel('residual', fontsize=25)
```

Here we see the error is largest at the edges. I think that is dominated by the first order finite difference in calculating the residual.
If I increase the number of integration points the error at the ede should drop.

```python
def f(r, k, m, b_z, b_theta):
    return r*(k*r*b_z + m*b_theta)**2/(k**2*r**2 + m**2)

def g(r, k, m, b_z, b_theta, p_prime, beta_0):
    term1 = beta_0*k**2*r**2/(k**2*r**2 + m**2)*p_prime
    term2 = r*(k*b_z)**2*(k**2*r**2 + m**2-1.)/(k**2*r**2 + m**2)
    term3 = 1/r*(m*b_theta)**2*(k**2*r**2+m**2-1.)/(k**2*r**2 + m**2)
    term4 = 2.*m*k*r*b_z*b_theta*(k**2*r**2 + m**2-1.)/(k**2*r**2 + m**2)
    term5 = 2*k**4*r**3*b_z**2/(k**2*r**2 + m**2)**2
    term6 = -2*k**2*m**2*r*b_theta**2/(k**2*r**2 + m**2)**2
    return term1 + term2 + term3 + term4 + term5 + term6

def der(r, y, params):
    f_params = {'r': r, 'k': params['k'], 'm': params['m'], 'b_z': params['b_z'](r),
                'b_theta': params['b_theta'](r)}
    g_params = {'r': r, 'k': params['k'], 'm': params['m'], 'b_z': params['b_z'](r),
                'b_theta': params['b_theta'](r), 'p_prime': params['p_prime'](r), 'beta_0': params['beta_0']}
    y_der = np.zeros(2)
    y_der[0] = y[1]/f(**f_params)
    y_der[1] = g(**g_params)*y[0]
    return y_der
```

```python
core_radius_norm = 0.7
transition_width_norm = 0.1
skin_width_norm = 0.1
profile = es.UnitlessSmoothedCoreSkin(lambda_bar=lambda_bar, k_bar=k_bar, epsilon=epsilon, points_core=5000, points_transition=5000, 
                                      points_skin=5000, core_radius_norm=core_radius_norm, 
                                      transition_width_norm=transition_width_norm, skin_width_norm=skin_width_norm)
params = {'m': 0., 'k': profile.k_bar, 'r_0': 0., 'a': 1., 'b': 'infinity', 'beta_0': profile.get_splines()['beta'](0)}
params.update(profile.get_splines())
params.update({'xi_factor': 1., 'magnetic_potential_energy_ratio': 1.})
beta_0 = profile.get_splines()['beta'](0)
```

```python
def f(r, k, m, b_z, b_theta):
    return r*(k*r*b_z + m*b_theta)**2/(k**2*r**2 + m**2)

def g(r, k, m, b_z, b_theta, p_prime, beta_0):
    term1 = beta_0*k**2*r**2/(k**2*r**2 + m**2)*p_prime
    term2 = r*(k*b_z)**2*(k**2*r**2 + m**2-1.)/(k**2*r**2 + m**2)
    term3 = 1/r*(m*b_theta)**2*(k**2*r**2+m**2-1.)/(k**2*r**2 + m**2)
    term4 = 2.*m*k*r*b_z*b_theta*(k**2*r**2 + m**2-1.)/(k**2*r**2 + m**2)
    term5 = 2*k**4*r**3*b_z**2/(k**2*r**2 + m**2)**2
    term6 = -2*k**2*m**2*r*b_theta**2/(k**2*r**2 + m**2)**2
    return term1 + term2 + term3 + term4 + term5 + term6

def der(r, y, params):
    f_params = {'r': r, 'k': params['k'], 'm': params['m'], 'b_z': params['b_z'](r),
                'b_theta': params['b_theta'](r)}
    g_params = {'r': r, 'k': params['k'], 'm': params['m'], 'b_z': params['b_z'](r),
                'b_theta': params['b_theta'](r), 'p_prime': params['p_prime'](r), 'beta_0': beta_0}
    y_der = np.zeros(2)
    y_der[0] = y[1]/f(**f_params)
    y_der[1] = g(**g_params)*y[0]
    return y_der
```

```python
r_init = 1E-3
r1 = np.linspace(r_init, core_radius_norm, 10000)
r2 = np.linspace(core_radius_norm, core_radius_norm + transition_width_norm, 10000)
r3 = np.linspace(core_radius_norm + transition_width_norm, core_radius_norm + transition_width_norm + skin_width_norm, 10000)
r4 = np.linspace(core_radius_norm + transition_width_norm + skin_width_norm, 1., 10000)

integrator = ode(der)
integrator.set_integrator('lsoda')
integrator.set_f_params(params)
f_params = {'r': r_init, 'k': params['k'], 'm': params['m'], 'b_z': params['b_z'](r_init),
            'b_theta': params['b_theta'](r_init)}
init = [r_init, 1.*f(**f_params)]
integrator.set_initial_value(init, t=r_init)
results1 = np.zeros((2, r1.size))
results1[:,0] = init
for i, position in enumerate(r1[1:]):
    integrator.integrate(position)
    results1[:, i+1] = integrator.y

integrator = ode(der)
integrator.set_integrator('lsoda')
integrator.set_f_params(params)
r_init = r2[0]
init = [results1[0, -1], results1[1, -1]]
integrator.set_initial_value(init, t=r_init)
results2 = np.zeros((2, r2.size))
results2[:,0] = init    
for i, position in enumerate(r2[1:]):
    integrator.integrate(position)
    results2[:, i+1] = integrator.y

integrator = ode(der)
integrator.set_integrator('lsoda')
integrator.set_f_params(params)
r_init = r3[0]
init = [results2[0, -1], results2[1, -1]]
integrator.set_initial_value(init, t=r_init)
results3 = np.zeros((2, r3.size))
results3[:,0] = init     
for i, position in enumerate(r3[1:]):
    integrator.integrate(position)
    results3[:, i+1] = integrator.y

integrator = ode(der)
integrator.set_integrator('lsoda')
integrator.set_f_params(params)
r_init = r4[0]
init = [results3[0, -1], results3[1, -1]]
integrator.set_initial_value(init, t=r_init)
results4 = np.zeros((2, r4.size))
results4[:,0] = init         
for i, position in enumerate(r4[1:]):
    integrator.integrate(position)
    results4[:, i+1] = integrator.y
```

```python
xi_array = np.append(np.append(np.append(results1[0], results2[0]), results3[0]), results4[0])
xi_der_f = np.append(np.append(np.append(results1[1], results2[1]), results3[1]), results4[1])
r_array = np.append(np.append(np.append(r1, r2), r3), r4)
```

```python
dr = np.gradient(r_array)
f_params = {'r': r_array, 'k': params['k'], 'm': params['m'], 'b_z': params['b_z'](r_array),
            'b_theta': params['b_theta'](r_array)}
g_params = {'r': r_array, 'k': params['k'], 'm': params['m'], 'b_z': params['b_z'](r_array),
            'b_theta': params['b_theta'](r_array), 'p_prime': params['p_prime'](r_array), 'beta_0': beta_0}
residual = f(**f_params)*np.gradient(xi_der_f/f(**f_params))/dr + np.gradient(f(**f_params))/dr*xi_der_f/f(**f_params) - g(**g_params)*xi_array
```

```python
plt.plot(r_array, np.abs(residual))
plt.yscale('log')
plt.xlabel(r'r')
plt.ylabel(r'residual')
```
