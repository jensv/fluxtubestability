In this notebook I will scan the kink stability map from Wesson's tokamak book with my newcomb code.
Wesson shows the stability boundaries for internal and external modes for a given axial current profile of the form:
$$j_{z}=j_{0}(1-(r/a)^{2})^{\nu}$$
The plot is shown below:

```python
from IPython.display import Image
Image('Wesson - stability map.jpg', width=500, height=500)
```

To recreate this map I will create equilbira for integer $\nu$ and scan increasing profiles with $q_{a}$ for m=1 and m=-1 modes while varying k between 1 and 100.
I will note any internal and external instabilities and mark them with dots on the map.

```python
def scan_along_constant_nu(nu, k_spacing, qa_spacing, offset=1E-4, mu_0=1., printer=True):
    r"""
    Scan profiles with a given nu but varying q on axis for stability agianst m=1 & m=-1 and a range of k values modes.
    
    Parameters
    ----------
    nu: float
        Given nu of profile. Nu is the exponent in the axial current porfile equation.    
    k_spacing: array_like
        Inputs to np.linspace giving the range of k to scan.
    qa_spacing: array_like
        Inputs to np.linspace giving the range of q on axis to scan.
    offset: float
        offset around singlarities in which to avoid integration. 
        A Frobenius solution is calulated at the offset.
        
    Returns
    -------
    points_x: ndarrary of floats (M)
        scanned q on axis value, form the x-axis of the Wesson stability map        
    internal_unstable_x: ndarray of floats (M)
        values of q on axis where an internal instability was detected for one of the m=1, m=-1, k number modes.
    external_unstale_x: ndarray of floats (M)
        values of q on axis where an external instability was detected for one of the m=1, m=-1, k number modes.
    
    Notes
    -----
    NuCurrentProfile constructor determines the symbolic equations for any nu prfoile. The actual profiles for a 
    given q on axis are calculated in NuCurrentProfile. Q on axis determines the j0 used in the nu curren tprofile equation.
    """
    print('start scanning nu =', nu)
    sing_search_points=1000
    dr = np.linspace(0, 1, 250)[1]
    k_points = np.linspace(k_spacing[0], k_spacing[1], num=k_spacing[2])
    qa_points = np.logspace(qa_spacing[0], qa_spacing[1], num=qa_spacing[2])
    
    profile_constructor = es.NuCurrentConstructor(nu=nu)
    
    points_x = set([])
    internal_unstable_x = set([])
    external_unstable_x = set([])
    
    missing_sing_cases = []
    
    for qa in qa_points:
        if printer:
            print("scanning qa =", qa,", for m = 1 & m = -1 and", k_spacing[2], "k values between", k_spacing[0], "and", k_spacing[1])
        for k in k_points:
            for m in [-1., 1.]:
                
                points_x.add(qa)
                
                params = {'mu_0': mu_0, 'k': k, 'm': m, 'r_0': 0., 'a': 1., 'b': 'infinity'}
                profile = es.NuCurrentProfile(nu_constructor=profile_constructor, nu=nu, qa=qa, k=1., mu_0=mu_0)
                params.update(profile.get_splines())
                
                (stable_internal, stable_external, 
                 xi, xi_der, r_array, delta_w, missing_sing_case) = new.stability(dr, offset, sing_search_points, params, suppress_output=True)
                
                if len(missing_sing_case) > 0:
                    missing_sing_cases.append(missing_sing_case)
                
                if not stable_internal:
                    internal_unstable_x.add(qa)
                if not stable_external:
                    external_unstable_x.add(qa)
                
    return np.array(list(points_x)), np.array(list(internal_unstable_x)), np.array(list(external_unstable_x)), missing_sing_cases
```

```python
x, y, z, missing_sing_cases_nu_0 = scan_along_constant_nu(0, (1, 10., 20), (-0.12, 1., 40), 
                                                          mu_0=consts.mu_0, printer=False)
x, y, z, missing_sing_cases_nu_1 = scan_along_constant_nu(1, (1, 10., 20), (-0.12, 1., 40), 
                                                          mu_0=consts.mu_0, printer=False)
x, y, z, missing_sing_cases_nu_2 = scan_along_constant_nu(2, (1, 10., 20), (-0.12, 1., 40), 
                                                          mu_0=consts.mu_0, printer=False)
x, y, z, missing_sing_cases_nu_3 = scan_along_constant_nu(3, (1, 10., 20), (-0.12, 1., 40), 
                                                          mu_0=consts.mu_0, printer=False)
```
