In this notebook I will scan the kink stability map from Wesson's tokamak book with my newcomb code.
Wesson shows the stability boundaries for internal and external modes for a given axial current profile of the form:
$$j_{z}=j_{0}(1-(r/a)^{2})^{\nu}$$
The plot is shown below:

```python
from IPython.display import Image
Image('Wesson - stability map.jpg', width=500, height=500)
```

To recreate this map I will create equilbira for integer $\nu$ and scan increasing profiles with $q_{a}$ for m=1 and m=-1 modes while varying k between 1 and 100.
I will note any internal and external instabilities and mark them with dots on the map.

```python
def scan_along_constant_nu(nu, k_spacing, qa_spacing, offset=1E-2, suydam_offset=1E-1, mu_0=1., printer=True, 
                           k_profile=1., m=1., k_profile_as_mode=False, b_z0 = 1.):
    r"""
    Scan profiles with a given nu but varying q on axis for stability agianst m=1 & m=-1 and a range of k values modes.
    
    Parameters
    ----------
    nu: float
        Given nu of profile. Nu is the exponent in the axial current porfile equation.    
    k_spacing: array_like
        Inputs to np.linspace giving the range of k to scan.
    qa_spacing: array_like
        Inputs to np.linspace giving the range of q on axis to scan.
    offset: float
        offset around singlarities in which to avoid integration. 
        A Frobenius solution is calulated at the offset.
        
    Returns
    -------
    points_x: ndarrary of floats (M)
        scanned q on axis value, form the x-axis of the Wesson stability map        
    internal_unstable_x: ndarray of floats (M)
        values of q on axis where an internal instability was detected for one of the m=1, m=-1, k number modes.
    external_unstale_x: ndarray of floats (M)
        values of q on axis where an external instability was detected for one of the m=1, m=-1, k number modes.
    
    Notes
    -----
    NuCurrentProfile constructor determines the symbolic equations for any nu prfoile. The actual profiles for a 
    given q on axis are calculated in NuCurrentProfile. Q on axis determines the j0 used in the nu current profile equation.
    """
    print('start scanning nu =', nu)
    sing_search_points=1000
    dr = np.linspace(0, 1, 250)[1]
    k_points = np.linspace(k_spacing[0], k_spacing[1], num=k_spacing[2])
    qa_points = np.logspace(np.log10(qa_spacing[0]), np.log10(qa_spacing[1]), num=qa_spacing[2])
    
    profile_constructor = es.NuCurrentConstructor(nu=nu)
    
    points_x = set([])
    internal_unstable_x = set([])
    external_unstable_x = set([])
    
    missing_end_cases = []
    
    for qa in qa_points:
        if printer:
            print("scanning qa =", qa,", for m = 1 & m = -1 and", k_spacing[2], "k values between", k_spacing[0], "and", k_spacing[1])
        for k in k_points:
            for m in [-m, m]:
                
                points_x.add(qa)
                
                params = {'mu_0': mu_0, 'k': k, 'm': m, 'r_0': 0., 'a': 1., 'b': 'infinity'}
                if k_profile_as_mode:
                    profile = es.NuCurrentProfile(nu_constructor=profile_constructor, nu=nu, qa=qa, k=k, mu_0=mu_0, b_z0=b_z0)
                else:
                    profile = es.NuCurrentProfile(nu_constructor=profile_constructor, nu=nu, qa=qa, k=k_profile, mu_0=mu_0, b_z0=b_z0)
                params.update(profile.get_splines())
                
                
                (stable_internal, stable_external, 
                     xi, xi_der, r_array, delta_w, missing_end_params) = new.stability(dr, offset, suydam_offset, sing_search_points, params, suppress_output=True)
                
                if missing_end_params:
                    missing_end_cases.append(missing_end_params) 
                
                if not stable_internal:
                    internal_unstable_x.add(qa)
                if not stable_external:
                    external_unstable_x.add(qa)
                
    return np.array(list(points_x)), np.array(list(internal_unstable_x)), np.array(list(external_unstable_x)), missing_end_cases
```

```python
def compare_wesson_plot(nu_0_results, nu_1_results, nu_2_results, nu_3_results):
    
    points_x, internal_unstable_x_0, external_unstable_x_0, missing = nu_0_results
    points_x, internal_unstable_x_1, external_unstable_x_1, missing = nu_1_results
    points_x, internal_unstable_x_2, external_unstable_x_2, missing = nu_2_results
    points_x, internal_unstable_x_3, external_unstable_x_3, missing = nu_3_results
    
    
    axes = host_subplot(111)

    axes.plot([8., 7., 6., 5., 4., 3., 2., 1., 0.5], [8., 7., 6., 5., 4., 3., 2., 1., 0.5], '-', c='grey', alpha=0.5)

    verts = [(8., 8.), (5., 5.), (0.5,0.5), (0.5,8.)]
    internal_poly = Polygon(verts, facecolor=None, fill=False, edgecolor='k', hatch='\\', alpha=0.5)
    internal_hatch = axes.add_patch(internal_poly)


    verts = [(10., 2.), (2., 2.), (2., 3.5655), (3./2. ,2.), (3./2., 2.8219), (4./3.+0.01, 2.)] + [(1., 2.), (1., 8.), (0.75,8.), (0.75,0.75), (10., 0.75)]
    external_poly = Polygon(verts, facecolor=None, fill=False, edgecolor='k', hatch='/', alpha=0.5)
    external_hatch = axes.add_patch(external_poly)

    axes.plot([1, 1, 4./3.+0.01, 3./2., 3./2., 2., 2.,10], [8, 2., 2., 2.8219, 2., 3.5655, 2, 2], '-', c='grey', alpha=0.5)


    point = axes.scatter(points_x, np.ones(points_x.size), marker='o', s=50, c=sns.xkcd_rgb['black'], linewidth=0)
    internal = axes.scatter(internal_unstable_x_0, np.ones(internal_unstable_x_0.size), marker='o', s=100, c=sns.xkcd_rgb['green'], linewidths=0)
    external = axes.scatter(external_unstable_x_0, np.ones(external_unstable_x_0.size), marker='o', s=50, c=sns.xkcd_rgb['pale red'], linewidth=0)

    axes.scatter(points_x, np.ones(points_x.size)*2, marker='o', s=50, c=sns.xkcd_rgb['black'], linewidth=0)
    axes.scatter(internal_unstable_x_1, np.ones(internal_unstable_x_1.size)*2, marker='o', s=100, c=sns.xkcd_rgb['green'], linewidths=0)
    axes.scatter(external_unstable_x_1, np.ones(external_unstable_x_1.size)*2, marker='o', s=50, c=sns.xkcd_rgb['pale red'], linewidth=0)

    axes.scatter(points_x, np.ones(points_x.size)*3, marker='o', s=50, c=sns.xkcd_rgb['black'], linewidth=0)
    axes.scatter(internal_unstable_x_2, np.ones(internal_unstable_x_2.size)*3, marker='o', s=100, c=sns.xkcd_rgb['green'], linewidths=0)
    axes.scatter(external_unstable_x_2, np.ones(external_unstable_x_2.size)*3, marker='o', s=50, c=sns.xkcd_rgb['pale red'], linewidth=0)

    axes.scatter(points_x, np.ones(points_x.size)*4, marker='o', s=50, c=sns.xkcd_rgb['black'], linewidth=0)
    axes.scatter(internal_unstable_x_3, np.ones(internal_unstable_x_3.size)*4, marker='o', s=100, c=sns.xkcd_rgb['green'], linewidths=0)
    axes.scatter(external_unstable_x_3, np.ones(external_unstable_x_3.size)*4, marker='o', s=50, c=sns.xkcd_rgb['pale red'], linewidth=0)


    legend = axes.legend([point, internal, external, internal_hatch, external_hatch], ['scanned point', r'internal $m=1$ instability', 
                                                                                   r'external $m=1$ instability', 'Wesson internallly unstable', 
                                                                                   'Wesson externally unstable'], 
    fancybox=True, loc='best', bbox_to_anchor=(1.1, 1), frameon=True, framealpha=1)

    legend_frame = legend.get_frame()
    legend_frame.set_linewidth(2)

    axes2 = axes.twin()
    axes2.set_xticks([8./7., 7./6., 6./5., 5./4., 4./3., 3./2., 2, 7./3., 5./2., 8./3., 3., 7./2., 4., 5, 6, 7, 8])
    axes2.set_xticklabels(["7\n8", "6\n7", "5\n6", "4\n5", "3\n4", "2\n3", "1\n2", "3\n7", "2\n5", "3\n8", "1\n3",
                           "2\n7", "1\n4", "1\n5", "1\n6", "1\n7", "1\n8"], fontsize=11)
    axes2.set_xlabel(r"Mode numbers $\binom{n}{m}$", labelpad=20)
    axes2.yaxis.tick_left()
    axes2.set_yticks([1, 2, 3, 4])
    axes2.set_yticklabels(["1", "2", "3", "4"])
    axes.set_ylabel(r"$\frac{q_{a}}{q_{o}}$", labelpad=50, rotation='horizontal', fontsize=30)
    axes2.grid(b=False)


    axes.set_yscale('log')
    axes.set_xscale('log')
    axes.invert_xaxis()
    axes.invert_yaxis()
    axes.set_xlim(10, 0.76)
    axes.set_ylim(8, 0.9)
    axes.set_xticks([1, 2, 3, 4, 5, 6, 7, 8])
    axes.set_yticks([1, 2, 3, 4])
    axes.set_xticklabels(["1", "2", "3", "4", "5", "6", "7", "8"])
    axes.set_yticklabels([r"$\nu = 0$", r"$\nu = 1$", r"$\nu = 2$", r"$\nu = 3$"])
    axes2.set_ylabel(r"$j_{z}$ profiles", rotation="horizontal", labelpad=125, fontsize = 22)
    axes.yaxis.tick_right()
    axes.set_xlabel(r'$q_{a}$', fontsize=22)

    plt.savefig('wesson_data.pdf')
    plt.show()
    #plt.savefig('wesson-stability.png')
```
