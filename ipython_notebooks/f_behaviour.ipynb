# Looking at the behaviour of the f numerator i.e. singularties of the Euler-Lagrange equation

The Euler-Lagrange equation is singular everywhere f is zero. So finding singularties of the Euler-Lagrange equation boils down to finding where f is 0. 

## Setup

### Setup plotting

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
sns.set_context('poster')
%matplotlib inline
```

```python
import numpy as np
```

### Setup f denominator

```python
def f_denom(r, k, m, b_z, b_theta):
    r"""
    Returns the relevant part of the f denominator.
    """
    return k*r*b_z + m*b_theta
```

## First look at constant pressure profiles from Newcomb's paper

```python
def f_behaviour_constant_pressure(k=1., m=-1., a=0.1, r_0i=1.0, b_z0=0.1, b_thetai=0.1):
    """
    Plot f for a constant pressure profile as described in Newcomb's 1960 paper.
    """
    profiles = es.NewcombConstantPressure()
    splines = profiles.get_splines()
    r = np.linspace(a, r_0i, 1000)
    f = f_denom(r, k, m, splines['b_z'](r), splines['b_theta'](r)) 
    plt.plot(r, f)
    sns.axlabel(r'$r$', r'$f$', fontsize=20)
    plt.title('$k=$'+str(k)+', $m=$'+str(m), fontsize=20)
```

## Parabolic pressure profile

```python
def f_behaviour_parabolic(k=1, m=-1, a=1.0, b_z0=0.1, b_thetai=0.1, q0=1.0):
    """
    Plot f for a parabolic current profile.
    """
    profiles = es.ParabolicNu2(q0=q0, k=k, b_z0=b_z0, a=1.0)
    splines = profiles.get_splines()
    r = np.linspace(0., a, 1000)
    f = f_denom(r, k, m, splines['b_z'](r), splines['b_theta'](r)) 
    plt.plot(r, f)
    sns.axlabel(r'$r$', r'$f$', fontsize=20)
    plt.title('$k=$'+str(k)+', $m=$'+str(m), fontsize=20)
```

## Skin core current profile

```python
def f_behaviour_skin_core(k=1, m=-1, a=1.0, b=1.0, j_skin=0.1, j_core=0.1, delta=0.5, b_z=0.1, t=0.2):
    """
    Plot f for a skin and core current profile.
    """
    mu0 = 1.
    n = m
    major_r_min = 1.0
    major_r_max = 10.0
    transition = t 
    pinch = {'a': a, 'b': b, 'j_skin': j_skin, 'j_core': j_core, 'delta': delta, 'b_z': b_z, 'mu0': mu0, 'k': k, 
             'n': n, 'major_r_min': major_r_min, 'major_r_max': major_r_max ,'transition': transition, 'smooth': True}
    j_dict = equil.define(pinch)
    r = np.linspace(0, 1, 1000)
    b_theta = equil.b_theta(r, j_dict, initial=0.0)
    f = f_denom(r, k, m, pinch['b_z'], b_theta) 
    plt.plot(r, f)
    sns.axlabel(r'$r$', r'$f$', fontsize=20)
    plt.title('$k=$'+str(k)+', $m=$'+str(m), fontsize=20)
```

```python
f_behaviour_skin_core()
```

```python
f_behaviour_skin_core(j_skin=1, j_core=0.5, b_z=0.1)
```
